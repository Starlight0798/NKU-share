```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# 
plt.style.use('seaborn-whitegrid')

# matplotlib
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

# 
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 
plt.figure(figsize=(10, 6), facecolor='white')

# 
plt.plot(x, y1, label='', color='tab:blue', linestyle='-', marker='o', markersize=6, linewidth=2)

# 
plt.plot(x, y2, label='', color='tab:green', linestyle='--', marker='^', markersize=6, linewidth=2)

# 
plt.title('', fontsize=16, fontweight='bold')
plt.legend(loc='best', fontsize=12)

# 
plt.xlabel('', fontsize=14)
plt.ylabel('', fontsize=14)

# 
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(0.1))

# 
for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_color('grey')
    spine.set_linewidth(0.5)

# 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

save_path = 'D:\\study\\\\\\\\3-4\\assets'

# Data
loss_rates = np.array([0, 2, 5, 10, 15])
delays = np.array([0, 50, 150, 300, 500])
window_sizes = np.array([4, 8, 16, 24, 32])
```

```python
# 

# 1.50ms
time_and_wait_times_50ms = np.array([3.90, 4.59, 5.21, 6.88, 10.13])
gbn_times_50ms = np.array([2.98, 6.47, 8.04, 10.22, 17.20])
sr_times_50ms = np.array([2.96, 5.48, 6.25, 6.00, 8.13])

plt.figure(figsize=(10, 6))
plt.style.use('seaborn-darkgrid')
plt.plot(loss_rates, time_and_wait_times_50ms, marker='o', linestyle='-', linewidth=2, markersize=8, label='Time-And-Wait')
plt.plot(loss_rates, gbn_times_50ms, marker='s', linestyle='--', linewidth=2, markersize=8, label='GBN')
plt.plot(loss_rates, sr_times_50ms, marker='d', linestyle='-.', linewidth=2, markersize=8, label='SR')

plt.title('Fixed Delay (50ms) with Varying Loss Rates', fontsize=16, fontweight='bold')
plt.xlabel('Loss Rate (%)')
plt.ylabel('Time (s)')
plt.legend(loc='best', fontsize=12)
plt.grid(True)
plt.tight_layout()
path = os.path.join(save_path, '1-1.jpg')
plt.savefig(path, bbox_inches='tight')
plt.show()


throughput_stop_wait = [475.51, 404.30, 355.88, 270.16, 183.44]
throughput_gbn = [622.23, 287.12, 230.81, 181.77, 107.97]
throughput_sr = [625.58, 338.69, 297.18, 309.56, 228.60]

plt.figure(figsize=(10, 6))
plt.plot(loss_rates, throughput_stop_wait, marker='o', linewidth=2, markersize=8, label='Time-and-Wait')
plt.plot(loss_rates, throughput_gbn, marker='s', linewidth=2, markersize=8, label='GBN')
plt.plot(loss_rates, throughput_sr, marker='^', linewidth=2, markersize=8, label='SR')
plt.title('Throughput vs. Loss Rate (Fixed Delay: 50ms)', fontsize=16, fontweight='bold')
plt.xlabel('Loss Rate (%)')
plt.ylabel('Throughput (kbps)')
plt.legend(loc='best', fontsize=12)
plt.grid(True)
plt.tight_layout()
path = os.path.join(save_path, '1-2.jpg')
plt.savefig(path, bbox_inches='tight')
plt.show()

# 2.2%
time_and_wait_times_2loss = np.array([3.56, 4.59, 6.08, 8.70, 13.50])
gbn_times_2loss = np.array([5.80, 6.47, 9.70, 12.19, 17.05])
sr_times_2loss = np.array([3.92, 3.94, 5.72, 6.47, 8.72])

plt.figure(figsize=(10, 6))
plt.plot(delays, time_and_wait_times_2loss, marker='o', linestyle='-', linewidth=2, markersize=8, label='Time-And-Wait')
plt.plot(delays, gbn_times_2loss, marker='s', linestyle='--', linewidth=2, markersize=8, label='GBN')
plt.plot(delays, sr_times_2loss, marker='d', linestyle='-.', linewidth=2, markersize=8, label='SR')

plt.title('Fixed Loss Rate (2%) with Varying Delays', fontsize=16, fontweight='bold')
plt.xlabel('Delay (ms)')
plt.ylabel('Time (s)')
plt.legend(loc='best', fontsize=12)
plt.grid(True)
plt.tight_layout()
path = os.path.join(save_path, '1-3.jpg')
plt.savefig(path, bbox_inches='tight')
plt.show()


throughput_stop_wait_2loss = [521.29, 404.30, 305.59, 213.42, 137.58]
throughput_gbn_2loss = [320.45, 287.12, 191.42, 152.40, 108.96]
throughput_sr_2loss = [473.57, 338.69, 324.77, 287.12, 213.02]

plt.figure(figsize=(10, 6))
plt.plot(delays, throughput_stop_wait_2loss, marker='o', linewidth=2, markersize=8, label='Time-and-Wait')
plt.plot(delays, throughput_gbn_2loss, marker='s', linewidth=2, markersize=8, label='GBN')
plt.plot(delays, throughput_sr_2loss, marker='^', linewidth=2, markersize=8, label='SR')
plt.title('Fixed Loss Rate (2%) with Varying Delays', fontsize=16, fontweight='bold')
plt.xlabel('Delay (ms)')
plt.ylabel('Throughput (kbps)')
plt.legend(loc='best', fontsize=12)
plt.grid(True)
plt.tight_layout()
path = os.path.join(save_path, '1-4.jpg')
plt.savefig(path, bbox_inches='tight')
plt.show()
```
